```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
income_df = pd.read_csv("canada_per_capita_income.csv", names=["year", "income_usd"], header=0)
```

```python
income_df.head()
```

```python
plt.scatter(x=income_df["year"], y=income_df["income_usd"], marker="+")
plt.xlabel("Year")
plt.ylabel("Income (USD)")
plt.title("Canada's per capita income (USD)")
```

```python
model = linear_model.LinearRegression()
```

```python
x_train, x_test, y_train, y_test=train_test_split(income_df[["year"]], income_df.income_usd, test_size=0.1)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict(income_df[["year"]])
```

```python
plt.scatter(x=income_df["year"], y=income_df["income_usd"], marker="+")
plt.xlabel("Year")
plt.ylabel("Income (USD)")
plt.title("Canada's per capita income (USD)")
plt.plot(income_df.year, model.predict(income_df[["year"]]), color="red")
```
