# Single Variable Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
prices=pd.DataFrame(data={'area':[2600,3000,3200,3600,4000], 'price':[550000,565000,610000,680000,725000]} )
```

```python
prices
```

```python
plt.scatter(x=prices.area, y=prices.price)
plt.xlabel('Area')
plt.ylabel('Price')
```

### Creating and feeding model

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(prices[["area"]], prices.price)
```

### Prediction

```python
reg.predict([[3000]])
```

```python
reg.coef_
# tg alfa
```

```python
reg.intercept_
# offset
```

### Multiple input prediction

```python
areas_to_predict=[100, 2000, 5000, 9000, 3000, 4500]
```

```python
reg.predict([[i] for i in areas_to_predict])
```

#### Input as DataFrame

```python
to_predict_df=pd.DataFrame({"area":areas_to_predict})
```

```python
reg.predict(to_predict_df)
```

### Visual repr of linear regression model

```python
plt.scatter(x=prices.area, y=prices.price)
plt.xlabel('Area')
plt.ylabel('Price')
plt.plot(prices.area, reg.predict(prices[["area"]]), color="purple")
```

# Multiple Variable Linear Regression



```python
hp=pd.read_csv("homeprices.csv")
```

```python
hp
```

```python
hp.dropna(inplace=True)
```

```python
mreg=linear_model.LinearRegression()
mreg.fit(hp[['area', 'bedrooms', 'age']], hp.price)
```

```python
mreg.coef_
mreg.intercept_
```

```python
mreg.predict([[3000,3,40]])
```

### Prediction (data as df)

```python
df_to_predict=pd.DataFrame({'area':[4000,5000,6000], 'bedrooms':[3,3,4], 'age':[1,2,3]})
```

```python
mreg.predict(df_to_predict)
```
